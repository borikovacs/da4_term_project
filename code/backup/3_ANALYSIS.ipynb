```python
import pandas as pd
import numpy as np
from linearmodels.panel import PanelOLS
import smtplib # for statsmodels context
```

```python
# Load the PREPROCESSED data
df = pd.read_csv('../data/clean/covid_and_pollution_final.csv')
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])
df = df.set_index(['city_name', 'date'])

#core_cols = ['pm25', 'workplace_closing', 'temperature', 'wind speed', 'precipitation']
core_cols = ['pm25', 'workplace_closing', 'temperature']
df_clean = df[core_cols].dropna()
```

```python
y = df_clean['pm25']
#X = df_clean[['workplace_closing', 'temperature', 'wind speed', 'precipitation']]
X = df_clean[['workplace_closing', 'temperature']]
```

```python
model = PanelOLS(y, X, entity_effects=True)
results = model.fit(cov_type='clustered', cluster_entity=True)

print(results)
print(f"Within-R2: {results.rsquared_within:.4f}")
```
